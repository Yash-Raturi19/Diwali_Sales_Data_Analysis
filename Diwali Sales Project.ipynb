# Recommended Graph Types (Seaborn + Matplotlib)

1. countplot() → Gender, Age Group, Marital Status, Status

2. barplot() → Product Category vs Amount, Zone vs Orders

3. histplot() → Amount distribution

4. boxplot() → Amount by Gender / Age Group

5. heatmap() → Zone × Product Category
